Vamos a hacer un análisis más profundo de estos outliers para entender qué sucedió en esos partidos específicos.


```python
# Calcular el IQR (Interquartile Range) para FTHG
Q1 = full_df['FTHG'].quantile(0.25)
Q3 = full_df['FTHG'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers = full_df[(full_df['FTHG'] < lower_bound) | (full_df['FTHG'] > upper_bound)]
print("Número de outliers encontrados:", len(outliers))
print(outliers[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTR']])
```

```python
# Visualización de los outliers en un boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='FTR', y='FTHG', data=full_df)
plt.scatter(x=outliers['FTR'], y=outliers['FTHG'], color='red', label='Outliers')
plt.title('Distribución de goles del equipo local y outliers')
plt.xlabel('Resultado Final (FTR)')
plt.ylabel('Goles del equipo local (FTHG)')
plt.legend()
plt.show()
```

A continuación, creo nuevas variables muy interesantes que no están incluidas en el conjunto de datos original. Partidos totales jugados de cada equipo, Total de goles a favor y en contra de cada equipo y la diferencia de goles de cada equipo

```python
# Contar el número de partidos jugados como local y visitante para cada equipo
home_matches = full_df['HomeTeam'].value_counts()
away_matches = full_df['AwayTeam'].value_counts()

# Sumar los partidos jugados como local y visitante para obtener el total de partidos jugados
total_matches = home_matches.add(away_matches, fill_value=0)

# Sumar los goles anotados como local y visitante para cada equipo
home_goals = full_df.groupby('HomeTeam')['FTHG'].sum().astype(int)
away_goals = full_df.groupby('AwayTeam')['FTAG'].sum().astype(int)

# Sumar los goles concedidos como local y visitante para cada equipo
home_conceded_goals = full_df.groupby('HomeTeam')['FTAG'].sum().astype(int)
away_conceded_goals = full_df.groupby('AwayTeam')['FTHG'].sum().astype(int)

# Calcular los totales de goles anotados y concedidos
total_goals = home_goals.add(away_goals, fill_value=0)
total_conceded_goals = home_conceded_goals.add(away_conceded_goals, fill_value=0)

# Calcular la diferencia entre goles a favor y goles en contra
goal_difference = total_goals - total_conceded_goals

# Crear DataFrame combinando la información de goles anotados y concedidos, la diferencia de goles,
# y el total de partidos jugados para cada equipo
team_goals_df = pd.DataFrame({
    'Team': total_goals.index,
    'Total Matches Played': total_matches.values,
    'Total Scored Goals': total_goals.values,
    'Total Conceded Goals': total_conceded_goals.values,
    'Goal Difference': goal_difference.values
})

# Mostrar el DataFrame combinado
print(team_goals_df)
```

Voy a crear una función para predecir el resultado del partido basándonos simplemente en la diferencia de goles de cada equipo:

```python
def predict_winner(home_team_diff, away_team_diff):
    if home_team_diff > away_team_diff:
        return "Local"
    elif home_team_diff < away_team_diff:
        return "Visitante"
    else:
        return "Empate"

# Predecir resultado del partido
home_team_diff = team_goals_df.loc[team_goals_df["Team"] == "Real Madrid", "Goal Difference"].values[0]
away_team_diff = team_goals_df.loc[team_goals_df["Team"] == "Barcelona", "Goal Difference"].values[0]

result = predict_winner(home_team_diff, away_team_diff)
print("El resultado del partido entre Real Madrid y Barcelona es:", result)
```

En el ejemplo anterior se hace una simulación de un partido entre el Real Madrid y el Barcelona donde el Barcelona es el ganador del encuentro gracias a la mayor diferencia de goles positiva con respecto al Real Madrid

Para hacerlo más realista, creo una función para predecir el resultado del partido basándonos en la diferencia de goles de cada equipo y con ventaja para el equipo local (he añadido un peso del un +30%)

```python
# Función para predecir el resultado del partido basándonos en la diferencai de goles de cada equipo y con ventaja para el equipo local (he añadido un peso del un +30%)
def predict_winner(home_team_diff, away_team_diff, advantage=0.3):
    adjusted_home_diff = home_team_diff + (home_team_diff * advantage)
    
    if adjusted_home_diff > away_team_diff:
        return "Local"
    elif adjusted_home_diff < away_team_diff:
        return "Visitante"
    else:
        return "Empate"

# Predecir resultado del partido
home_team_diff = team_goals_df.loc[team_goals_df["Team"] == "Real Madrid", "Goal Difference"].values[0]
away_team_diff = team_goals_df.loc[team_goals_df["Team"] == "Real Madrid", "Goal Difference"].values[0]

result = predict_winner(home_team_diff, away_team_diff)
print("El resultado del partido entre Real Madrid y Barcelona es:", result)
```

Vemos cómo gracias a añadir el peso de favoritismo hacia el equipo local, el resultado del partido cambia y en este caso es el Madrid el ganador del partido
